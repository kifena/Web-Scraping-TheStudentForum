# Import necessary modules

```python
import pandas as pd
from bs4 import BeautifulSoup 
from selenium import webdriver
import lxml
import lxml.html.soupparser
import re
import requests
#NotebookApp.iopub_data_rate_limit=1.0e10
```

```python
headers = {
        'accept-encoding': 'gzip, deflate, sdch, br',
        'accept-language': 'en-US,en;q=0.8,zh-CN;q=0.6,zh;q=0.4',
        'upgrade-insecure-requests': '1',
        'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36',
        'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
        'cache-control': 'max-age=0',
    }

base_url = ['https://www.thestudentroom.co.uk/forumdisplay.php?f=269',
            'https://www.thestudentroom.co.uk/forumdisplay.php?f=223',
            'https://www.thestudentroom.co.uk/forumdisplay.php?f=176',
            'https://www.thestudentroom.co.uk/forumdisplay.php?f=229',
            'https://www.thestudentroom.co.uk/forumdisplay.php?f=151',
            'https://www.thestudentroom.co.uk/forumdisplay.php?f=611',
            'https://www.thestudentroom.co.uk/forumdisplay.php?f=89',
            'https://www.thestudentroom.co.uk/forumdisplay.php?f=1538',
            'https://www.thestudentroom.co.uk/forumdisplay.php?f=362']

base_num = ['269','223','176','229','151','611','89','1538','362']

page_loop='&page='

#page_url = base_url+page_loop

post_links = []
k = -1
for base in base_url:
    page_url = base + page_loop
    k = k + 1
    for i in range(1,10):
        url = page_url + str(i)
        print('begin scrapping the link: ', url)
        request = requests.get(url, headers)
        soup = BeautifulSoup(request.text, 'lxml')
        body = soup.find('body')
        main_outlet = body.find('div',{'id':"page_forumdisplay"})
        try:
            all_a = main_outlet.findAll('table', {'class': "thread-list"})[0].findAll('a',{'data-fid':base_num[k]})
            all_a = all_a[2:-1]
            for a in all_a:
                if 'showthread.php' in a.get('href'):
                    post_link = a.get('href')
                    post_links.append(post_link)
                else:
                    continue
        except AttributeError:
            continue
```
